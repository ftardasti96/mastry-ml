```python
import numpy as np
from tensorflow import keras
```

```python
def generate_seq(length,feature):
    return [np.random.randint(0,feature) for _ in range(length)]
```

```python
def one_hot_encoding(seq,feature):
    encoding = []
    for i in seq:
        test = [0]*feature
        test[i] = 1
        encoding.append(test)
    return np.array(encoding)
```

```python
def one_hot_decoding(seq):
    decoding = []
    for i in seq:
        test = np.argmax(i)
        decoding.append(test)
    return(decoding)
```

```python
#define model
time_step = 25
feature = 10
Model = keras.models.Sequential()
Model.add(keras.layers.LSTM(32,input_shape=(time_step,feature))) #memory units, chosen with a little trial and error
Model.add(keras.layers.Dense(feature,activation='softmax'))
Model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['acc'])
Model.summary()
```

```python
def generate_example(length,feature,out_index):
    seq = generate_seq(length,feature)
    seq_encode = one_hot_encoding(seq,feature)
    X = seq_encode.reshape(1,length,feature)
    y = seq_encode[out_index].reshape(1,feature)
    return X,y
```

```python
for i in range(10000):
    X,y = generate_example(25,10,2)
    Model.fit(X,y,epochs=1)
```

```python
X,y = generate_example(25,100,2)
p = Model.predict(X)
```
