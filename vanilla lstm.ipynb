```python
import numpy as np
from tensorflow import keras
from tensorflow.keras.layers import Dense,LSTM
```

```python
def generate_sequence(length, n_features):
#     seq = []
#     for i in range(length):
#         seq.append(np.random.randint(0,n_features))
#     return seq
    return[np.random.randint(0,n_features) for _ in range(length)]
```

```python
# one hot encode sequence
def one_hot_encode(sequence, n_features):
    encoding = []
    for value in sequence:
        vector = [0 for _ in range(n_features)]
        vector[value] = 1
        encoding.append(vector)
    return np.array(encoding)
```

```python
# one hot decode sequence
def one_hot_decode(sequence):
    return [np.argmax(sequence[i]) for i in range(len(sequence))]
```

```python
def generate_example(length, n_features, out_index):
    seq = generate_sequence(length,n_features)
    seq_encode = one_hot_encode(seq,n_features)
    seq_decode = one_hot_decode(seq_encode)
    X = seq_encode.reshape(1, length, n_features)
    y = seq_encode[out_index, :].reshape(1, n_features)
    return X,y
```

```python
X,y = generate_example(5,10,2)
```

```python
# define model
length = 5
n_features = 10
out_index = 2
model = keras.models.Sequential()
model.add(LSTM(25, input_shape=(length, n_features))) #memory units, chosen with a little trial and error
model.add(Dense(n_features, activation= 'softmax' ))
model.compile(loss= 'categorical_crossentropy' , optimizer= 'adam' , metrics=[ 'acc' ])
print(model.summary())
```

```python
# fit model
for i in range(10000):
    X, y = generate_example(length, n_features, out_index)
    model.fit(X, y, epochs=1, verbose=2)
```

```python
# evaluate model
correct = 0
for i in range(100):
    X, y = generate_example(length, n_features, out_index)
    yhat = model.predict(X)
    if one_hot_decode(yhat) == one_hot_decode(y):
        correct += 1
print( 'Accuracy: %f' % ((correct/100)*100.0))
```
