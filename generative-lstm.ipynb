```python
import numpy as np
from tensorflow import keras
from matplotlib.patches import PathPatch
from matplotlib.path import Path
import matplotlib.pyplot as plt
```

```python
def random_rectangle():
    width,height = np.random.rand(),np.random.rand()
    point = list()
    point.append([0.0 , 0.0])
    point.append([width,0.0])
    point.append([width,height])
    point.append([0.0,height])
    return point
```

```python
rect = random_rectangle()
print(rect)
```

```python
def plot_rect(rect):
    rect.append(rect[0])
    code = [Path.MOVETO,
            Path.LINETO,
            Path.LINETO,
            Path.LINETO,
            Path.CLOSEPOLY]
    
    path = Path(rect, code)
    axis = plt.gca()
    patch = PathPatch(path,facecolor='orange')
    axis.add_patch(patch)
    plt.show()
```

```python
plot_rect(rect)
```

```python
def generate_sample():
    rect = random_rectangle()
    X,y = list(),list()
    for i in range(1,len(rect)):
        X.append(rect[i-1])
        y.append(rect[i])
    X = np.array(X)
    y = np.array(y)
    X = X.reshape(len(X),1,2)
    return X,y
```

```python
X, y = generate_sample()
for i in range(X.shape[0]):
    print(X[i][0], '=>' , y[i])
```

```python
#define model
model = keras.models.Sequential()
model.add(keras.layers.LSTM(10,input_shape=(1,2)))
model.add(keras.layers.Dense(2,activation='linear')) #dense is 2 bcz x and y coordination
model.compile(optimizer='adam',loss='mae')
model.summary()
```

```python
#fit model
for i in range(25000):
    X,y = generate_sample()
    model.fit(X,y,epochs=1,shuffle=False)
```

```python
def generate_rect(model):
    last = np.array([0,0]).reshape(1,1,2)
    rect = list()
    rect.append(last[0][0])
    for i in range(3):
        last = np.array(model.predict(last)).reshape(1,1,2)
        rect.append(last[0][0])
    plot_rect(rect)
```

```python
generate_rect(model)
```
